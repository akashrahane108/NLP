```python
#remove punctuations
import string
def rmv_punc(text):
  punct = string.punctuation
  return text.translate(str.maketrans('', '', punct))
```

```python
def return_numeric_char(text):
    text = re.sub('[^a-zA-Z0-9(/)]','',text)
    return text
```

```python
def return_char_remove_whitespace(text):
    text = re.sub('[^a-zA-Z0-9(/)]','',text)
    text = re.sub('\s+',' ', text)
    return text
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download("vader_lexicon")
```

```python
sentiments = SentimentIntensityAnalyzer()
df["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in df["summary"]]
df["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in df["summary"]]
df["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in df["summary"]]
df["Compound"] = [sentiments.polarity_scores(i)["compound"] for i in df["summary"]]
```

```python
df
```

```python
score = df["Compound"].values
```

```python
sentiment = []
for i in score:
    if i >= 0.05 :
        sentiment.append('Positive')
    elif i <= -0.05 :
        sentiment.append('Negative')
    else:
        sentiment.append('Neutral')
df["Sentiment"] = sentiment
```

```python
def sentiment_matched_for_different_columns(row):
    if (row["Summary_based_sentiment"] == row["Review_based_Sentiment"]) & (row["Review_based_Sentiment"]  == row["Rating_based_sentiment"]):
        return "Sentiments matched"
    else:
        return "Sentiments did not matched"
```
